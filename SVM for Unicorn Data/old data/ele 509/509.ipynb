```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_averages(n, num_experiments=100):
    """Simulate the averages for n samples of a Cauchy random variable."""
    averages = []
    for _ in range(num_experiments):
        samples = np.random.standard_cauchy(n)
        averages.append(np.mean(samples))
    return averages

def plot_histogram(n_values):
    """Plot histogram of sample averages for different n values."""
    plt.figure(figsize=(12, 6))

    colors = ['blue', 'orange', 'green', 'red']  # Corresponding colors for the n values
    for i, n in enumerate(n_values):
        averages = simulate_averages(n)
        plt.hist(averages, bins=50, alpha=0.6, color=colors[i], label=f'n={n}', range=[-40, 40])

    plt.title(' Q3. c: Distribution of Sample Means for Different Sample Sizes (Cauchy)')
    plt.xlabel('Sample Mean')
    plt.ylabel('Frequency')
    plt.legend(loc='upper right')
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.show()

# Different values of n
n_values = [100, 1000, 10000, 100000]

plot_histogram(n_values)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_gaussian_averages(mu, sigma, n, num_experiments=100):
    """Simulate the averages for n samples of a Gaussian random variable."""
    averages = []
    for _ in range(num_experiments):
        samples = np.random.normal(mu, sigma, n)
        averages.append(np.mean(samples))
    return averages

def plot_gaussian_histogram(mu, sigma, n_values):
    """Plot histogram of sample averages for different n values."""
    plt.figure(figsize=(12, 6))

    colors = ['blue', 'orange', 'green', 'red']  # Corresponding colors for the n values
    for i, n in enumerate(n_values):
        averages = simulate_gaussian_averages(mu, sigma, n)
        plt.hist(averages, bins=50, alpha=0.6, color=colors[i], label=f'n={n}', range=[-0.5, 0.5])

    plt.title('Q3. d Distribution of Sample Means for Different Sample Sizes (Gaussian)')
    plt.xlabel('Sample Mean')
    plt.ylabel('Frequency')
    plt.legend(loc='upper right')
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.xlim(-0.5, 0.5)  # Set x-axis limits
    plt.show()

# Parameters for Gaussian distribution
mu = 0  # mean
sigma = 1  # standard deviation

# Different values of n
n_values = [100, 1000, 10000, 100000]

plot_gaussian_histogram(mu, sigma, n_values)
```
